```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
```

```python
# load data
m = tf.keras.datasets.fashion_mnist
(x_train,y_train),(x_test,y_test) = m.load_data()
```

```python
class_name = ['t_shirt','trouser','pullover','dress','coat','sandal','shirt','sneaker','bag','ankle boot']
```

```python
x_test.shape
```

```python
x_train.shape
```

```python
plt.figure()
plt.imshow(x_train[1200])
plt.show()
```

```python
# scale the data
x_train = x_train/255.0
```

```python
# plot some simples
plt.figure(figsize=(10,10))
for i in range(25):
  plt.subplot(5,5,i+1)
  plt.grid(False)
  plt.xticks([])
  plt.yticks([])
  plt.imshow(x_train[i],cmap=plt.cm.binary)
  plt.xlabel(class_name[y_train[i]])
plt.show()
```

```python
# create the model
model=keras.Sequential()
model.add(keras.layers.Flatten(input_shape=(28,28)))
model.add(keras.layers.Dense(32,activation='relu'))
model.add(keras.layers.Dense(128,activation='relu'))
model.add(keras.layers.Dense(10,activation='softmax'))
```

```python
model.summary()
```

```python
# compile and train the model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(x_train,y_train,epochs=30)
```

```python
test_loss,test_acc = model.evaluate(x_test,y_test,verbose=2)
print('\n the test accuracy is ', test_acc)
```
